# 1.第一部分101改成1001

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt

def drawts(y,pname):
    ##draw ax
    fig = plt.figure(figsize=(10,8))
    ts_ax=plt.subplot2grid((2,2),(0,0),colspan=2)
    acf_ax=plt.subplot2grid((2,2),(1,0))
    pacf_ax=plt.subplot2grid((2,2),(1,1))
    ##draw plot
    ts_ax.plot(y,'*-')
    ts_ax.set_title('Time Series Analysis Plots')
    smt.graphics.plot_acf(y,lags=None,ax=acf_ax,alpha=0.05) ##2sigma
    smt.graphics.plot_pacf(y,lags=None,ax=pacf_ax,alpha=0.05)  ##2sigma
    #plt.savefig('%s.jpg'%pname,dpi=256)
    plt.show()
    plt.close()
```

生成$$ x_0 = 0 $$且$$ x_t = 0.8 x_{t-1} + \varepsilon_t $$的序列值$$ x_0,x_1,x_2,...,x_{100} $$

```python
x=np.zeros(1001)

for i in range(1,1001):
    x[i]=x[i-1]*0.8+np.random.normal(0, 1, 1)

drawts(x,'1')
```

生成$$ x_0 = 0 $$且$$ x_t = -0.8 x_{t-1} + \varepsilon_t $$的序列值$$ x_0,x_1,x_2,...,x_{100} $$

```python
x=np.zeros(1001)

for i in range(1,1001):
    x[i]=x[i-1]*-0.8+np.random.normal(0, 1, 1)

drawts(x,'1')
```

$$ x_t = x_{t-1} - 0.5 x_{t-2} + \varepsilon_t $$的序列值$$ x_0,x_1,x_2,...,x_{100} $$

```python
x=np.zeros(1001)
x[1]=np.random.normal(0, 1, 1)
x[2]=np.random.normal(0, 1, 1)

for i in range(2,1001):
    x[i]=x[i-1]-x[i-2]*0.5+np.random.normal(0, 1, 1)

drawts(x,'1')
```

$$ x_t = - x_{t-1} - 0.5 x_{t-2} + \varepsilon_t $$的序列值$$ x_0,x_1,x_2,...,x_{100} $$

```python
x=np.zeros(1001)
x[1]=np.random.normal(0, 1, 1)
x[2]=np.random.normal(0, 1, 1)

for i in range(2,1001):
    x[i]=-x[i-1]-x[i-2]*0.5+np.random.normal(0, 1, 1)

drawts(x,'1')
```
