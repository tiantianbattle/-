```python
import numpy as np
import matplotlib.pyplot as plt
```

生成$$ x_0 = 0 $$且$$ x_t = 0.8 x_{t-1} + \varepsilon_t $$的序列值$$ x_0,x_1,x_2,...,x_{100} $$

```python
x=np.zeros(101)

for i in range(1,101):
    x[i]=x[i-1]*0.8+np.random.normal(0, 1, 1)

plt.plot(x)
```

$$ x_0 = 0 $$且$$ x_t = -1.1 x_{t-1} + \varepsilon_t $$

```python
x=np.zeros(101)

for i in range(1,101):
    x[i]=-x[i-1]*1.1+np.random.normal(0, 1, 1)

plt.plot(x)
```

$$ x_0 和 x_1 均隨機生成 $$且$$ x_t = x_{t-1} - 0.5 x_{t-2} + \varepsilon_t $$

```python
x=np.zeros(101)
x[0]=np.random.normal(0, 1, 1)
x[1]=np.random.normal(0, 1, 1)

for i in range(1,101):
    x[i]=x[i-1]-0.5*x[i-2]+np.random.normal(0, 1, 1)

plt.plot(x)
```

$$ x_0 和 x_1 均隨機生成 $$且$$ x_t = x_{t-1} + 0.5 x_{t-2} + \varepsilon_t $$

```python
x=np.zeros(101)
x[0]=np.random.normal(0, 1, 1)
x[1]=np.random.normal(0, 1, 1)

for i in range(1,101):
    x[i]=x[i-1]+0.5*x[i-2]+np.random.normal(0, 1, 1)

plt.plot(x)
```
