```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.tsa.api as smt

def drawts(y,pname):
    ##draw ax
    fig = plt.figure(figsize=(10,8))
    ts_ax=plt.subplot2grid((2,2),(0,0),colspan=2)
    acf_ax=plt.subplot2grid((2,2),(1,0))
    pacf_ax=plt.subplot2grid((2,2),(1,1))
    ##draw plot
    ts_ax.plot(y,'*-')
    ts_ax.set_title('Time Series Analysis Plots')
    smt.graphics.plot_acf(y,lags=None,ax=acf_ax,alpha=0.05) ##2sigma
    smt.graphics.plot_pacf(y,lags=None,ax=pacf_ax,alpha=0.05)  ##2sigma
    #plt.savefig('%s.jpg'%pname,dpi=256)
    plt.show()
    plt.close()
```

考察$$ x_{t}-0.5x_{t-1}=\varepsilon_{t}-0.8\varepsilon_{t-1} $$

```python
n=1000
x=np.zeros(n+1)
ep=np.random.normal(0, 1, n+1)

for i in range(1,n):
    x[i]=x[i-1]*0.5+ep[i]-0.8*ep[i-1]

x=x[1:]
drawts(x,'x')
```

考察$$ x_{t}-0.5x_{t-1}=\varepsilon_{t}$$

```python
x1=np.zeros(n+1)

for i in range(1,n):
    x1[i]=x1[i-1]*0.5+ep[i]

x1=x1[1:]
drawts(x1,'x1')
```

考察$$ x_{t}=\varepsilon_{t}-0.8\varepsilon_{t-1} $$

```python
x2=np.zeros(n+1)

for i in range(1,n):
    x2[i]=ep[i]-0.8*ep[i-1]

x2=x2[1:]
drawts(x2,'x2')
```
