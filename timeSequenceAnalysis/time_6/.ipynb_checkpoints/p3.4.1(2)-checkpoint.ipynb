```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.stats.diagnostic import acorr_ljungbox as lb_test
import statsmodels.tsa.api as smt
```

# statsmodels自帶的畫TS/ACF/PACF的函數

```python
def drawts(y,pname):
    ##draw ax
    fig = plt.figure(figsize=(10,8))
    ts_ax=plt.subplot2grid((2,2),(0,0),colspan=2)
    acf_ax=plt.subplot2grid((2,2),(1,0))
    pacf_ax=plt.subplot2grid((2,2),(1,1))
    ##draw plot
    ts_ax.plot(y,'*-')
    ts_ax.set_title('Time Series Analysis Plots')
    smt.graphics.plot_acf(y,lags=10,ax=acf_ax,alpha=0.05) ##2sigma
    smt.graphics.plot_pacf(y,lags=10,ax=pacf_ax,alpha=0.05)  ##2sigma
    #plt.savefig('%s.jpg'%pname,dpi=256)
    plt.show()
    plt.close()
```

```python
#n=len of varepsilon
n=100

#varepsilon_t
ep=np.random.normal(0, 1, n)
```

# 1
<font color=#A52A2A size=5 >$$ x_{t}=\varepsilon_{t}-2\varepsilon_{t-1} $$</font>

```python
x=np.zeros(n)

for i in range(1,n):
    x[i]=ep[i]-2*ep[i-1]

x=x[1:]

drawts(x,'1')
```

# 2
<font color=#A52A2A size=5 >$$ x_{t}=\varepsilon_{t}-0.5\varepsilon_{t-1} $$</font>

```python
x=np.zeros(n)

for i in range(1,n):
    x[i]=ep[i]-0.5*ep[i-1]

x=x[1:]

drawts(x,'2')
```

# 3
<font color=#A52A2A size=5 >$$ x_{t}=\varepsilon_{t}-\frac{4}{5}\varepsilon_{t-1}+\frac{16}{25}\varepsilon_{t-2} $$</font>

```python
x=np.zeros(n)

for i in range(1,n):
    x[i]=ep[i]-0.8*ep[i-1]+0.64*ep[i-2]

x=x[2:]

drawts(x,'3')
```

# 4
<font color=#A52A2A size=5 >$$ x_{t}=\varepsilon_{t}-\frac{5}{4}\varepsilon_{t-1}+\frac{25}{16}\varepsilon_{t-2} $$</font>

```python
x=np.zeros(n)

for i in range(1,n):
    x[i]=ep[i]-1.25*ep[i-1]+1.5625*ep[i-2]

x=x[2:]

drawts(x,'4')
```
