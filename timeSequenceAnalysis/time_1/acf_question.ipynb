# 畫時序圖、延遲k自相關圖、延遲k偏自相關圖

```python
import statsmodels.tsa.api as smt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 利用statsmodels

```python
def drawts(y,pname):
    ##draw ax
    fig = plt.figure(figsize=(10,8))
    ts_ax=plt.subplot2grid((2,2),(0,0),colspan=2)
    acf_ax=plt.subplot2grid((2,2),(1,0))
    pacf_ax=plt.subplot2grid((2,2),(1,1))
    ##draw plot
    ts_ax.plot(y,'*-')
    ts_ax.set_title('Time Series Analysis Plots')
    smt.graphics.plot_acf(y,lags=None,ax=acf_ax,alpha=0.05) ##2sigma
    smt.graphics.plot_pacf(y,lags=None,ax=pacf_ax,alpha=0.05)  ##2sigma
    #plt.savefig('%s.jpg'%pname,dpi=256)
    plt.show()
    plt.close()
```

1、自协方差函数：$r(t,s)=E(X_t-\mu_t)(X_s-\mu_s)$<br>
2、延迟K自相关函数：$\rho_k=\frac{r(t,t+k)}{\sqrt{}{DX_t\cdot DX_{t+k}}}=\frac{r(k)}{\sigma_X^2}=\frac{r(k)}{r(0)}$<br>
3、标准差&两倍标准差：$\sqrt{{1/n}*(1+2*)}{2}$没补完

```python
def mydrawts(y,pname):
    myname='XIE_HAOMING'
    ##draw ax
    fig = plt.figure(figsize=(10,8))
    ts_ax=plt.subplot2grid((2,2),(0,0),colspan=2)
    acf_ax=plt.subplot2grid((2,2),(1,0))
    pacf_ax=plt.subplot2grid((2,2),(1,1))
    ##draw plot
    ts_ax.plot(y,'*-')
    ts_ax.set_title('Time Series Analysis Plots(custom %s)'%myname)
    
    ##calclate acf    
    myacf=np.ones((17))#用新的 rho r 来替换它
    #获取pname文件里的相关数据
    n= len(y)
    u = y.mean() #用u表示μ
    for k in range(0,17):
        Molecular = 0   #分子
        denominator = 0 #分母
        for i in range(0,n-k):
            Molecular += (y[i] - u)*(y[i+k] - u)
        for i in range(0,n):
            denominator += (y[i] - u)**2
        myacf[k] = Molecular/denominator

        #阴影表示的是两倍标准差
    twosigma=np.ones((17))
    num2= len(y)
    twosigma[0] = (1/num2) **0.5
    for i in range(1,17):
        sum = 0
        for j in range(0,i):
            sum +=myacf[j] **2
        twosigma[i] = ((1/num2)*(1+2*sum ))**0.5
    
    acf_ax.bar(range(len(myacf)),myacf)
    acf_ax.fill_between(range(len(myacf)),-1*twosigma,twosigma,color='lightblue')
    
    #plt.savefig('%s.jpg'%pname,dpi=256)
    plt.show()
    plt.close()
```

```python
##read data
dfname='附录1.2'
y=pd.read_csv('%s.csv'%dfname,header=None)
y.iloc[:,0]=y.iloc[:,0].astype('float')
y=y.values[:,0]
drawts(y,dfname)
mydrawts(y,dfname)
```
